## Modelo de Regresi칩n Lineal Simple

El modelo de regresi칩n lineal simple es una t칠cnica estad칤stica utilizada para modelar la relaci칩n entre dos variables: una variable dependiente (respuesta) y una variable independiente (predictora). La relaci칩n entre estas variables se representa mediante una l칤nea recta.

### Ecuaci칩n del Modelo

La ecuaci칩n del modelo de regresi칩n lineal simple es:

$ y = \beta_0 + \beta_1 x + \epsilon $

donde:
- $ y $ es la variable dependiente.
- $ x $ es la variable independiente.
- $ \beta_0 $ es la intersecci칩n con el eje $ y $ (constante o t칠rmino independiente).
- $ \beta_1 $ es la pendiente de la l칤nea de regresi칩n, que representa el cambio en $ y $ por unidad de cambio en $ x $.
- $ \epsilon $ es el t칠rmino de error, que captura la variabilidad en $ y $ que no se explica por la relaci칩n lineal con $ x $.

El prop칩sito del modelo de regresi칩n lineal simple es encontrar los valores de $ \beta_0 $ y $ \beta_1 $ que minimicen la suma de los cuadrados de las diferencias entre los valores observados y los valores predichos por el modelo. Este proceso se conoce como ajuste de m칤nimos cuadrados ordinarios.

#### Estimacion de coeficientes

$\hat{\beta}_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}$

$\hat{\beta}_0 = \bar{y} - \hat{\beta}_1 \bar{x}$

#### Varianza del error

La varianza del error ($\sigma^2$) mide cu치n dispersos est치n los valores de $y$ alrededor de la l칤nea de regresi칩n. Se utiliza para cuantificar la variabilidad de los valores de $洧녽$ que no se explica por el modelo de regresi칩n lineal simple.

$\sigma^2 = \frac{\sum (y_i - \hat{y}_i)^2}{n - 2}$

#### Error estandar

El error est치ndar de un coeficiente mide la precisi칩n con la que se estima el coeficiente. Menor error estandar indica una mayor precision en la estimacion, mientras que un error alto indica una menor precisi칩n en la estimaci칩n

$\text{SE}(\hat{\beta}_1) = \sqrt{\frac{\sigma^2}{\sum (x_i - \bar{x})^2}}$

$\text{SE}(\hat{\beta}_0) = \sqrt{\sigma^2 \left( \frac{1}{n} + \frac{\bar{x}^2}{\sum (x_i - \bar{x})^2} \right)}$


### Supuestos

Para que el modelo de regresi칩n lineal simple sea v치lido, se deben cumplir ciertas asunciones:
1. **Linealidad:** La relaci칩n entre la variable dependiente y la variable independiente es lineal.
2. **Independencia:** Las observaciones son independientes entre s칤. En otras palabras, que no exista autocorrelaci칩n
3. **Homoscedasticidad:** La variabilidad de los errores es constante a lo largo de todos los valores de la variable independiente.
4. **Normalidad de los Errores:** Los errores de la regresi칩n son distribuidos normalmente.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
tama침o = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1, 1)  # Convertimos a formato de columna
precio = np.array([200, 250, 300, 350, 400])

# Crear el modelo de regresi칩n lineal
modelo = LinearRegression()

# Ajustar el modelo a los datos
modelo.fit(tama침o, precio)

# Obtener los par치metros del modelo
beta_0 = modelo.intercept_
beta_1 = modelo.coef_[0]

# Imprimir los par치metros
print(f"Intersecci칩n (0): {beta_0:.2f}")
print(f"Pendiente (1): {beta_1:.2f}")

# Predicciones
tama침o_pred = np.linspace(900, 3100, 100).reshape(-1, 1)
precio_pred = modelo.predict(tama침o_pred)

# Visualizaci칩n
plt.figure(figsize=(10, 6))
plt.scatter(tama침o, precio, color='blue', label='Datos Observados')
plt.plot(tama침o_pred, precio_pred, color='red', linewidth=2, label='L칤nea de Regresi칩n')
plt.xlabel('Tama침o (pies cuadrados)')
plt.ylabel('Precio (miles de d칩lares)')
plt.title('Regresi칩n Lineal Simple')
plt.legend()
plt.grid(True)
plt.show()
```

## Modelo de Regresi칩n Lineal M칰ltiple

El modelo de regresi칩n lineal m칰ltiple es una extensi칩n del modelo de regresi칩n lineal simple que se utiliza para modelar la relaci칩n entre una variable dependiente y dos o m치s variables independientes. Este modelo permite examinar c칩mo varias variables predictoras influyen en una variable respuesta.

### Ecuaci칩n del Modelo

La ecuaci칩n del modelo de regresi칩n lineal m칰ltiple es:

$$ y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_n x_n + \epsilon $$

donde:
- $y$ es la variable dependiente.
- $x_1, x_2, \ldots, x_n$ son las variables independientes (predictoras).
- $\beta_0$ es la intersecci칩n con el eje $y$ (constante o t칠rmino independiente).
- $\beta_1, \beta_2, \ldots, \beta_n$ son los coeficientes de las variables independientes, que representan el cambio en $y$ por unidad de cambio en cada $x_i$.
- $\epsilon$ es el t칠rmino de error, que captura la variabilidad en $y$ que no se explica por la relaci칩n lineal con las $x_i$.

El prop칩sito del modelo de regresi칩n lineal m칰ltiple es encontrar los valores de $\beta_0, \beta_1, \ldots, \beta_n$ que minimicen la suma de los cuadrados de las diferencias entre los valores observados y los valores predichos por el modelo. Este proceso se realiza utilizando el ajuste de m칤nimos cuadrados ordinarios.

#### Estimacion de coeficientes 

En regresi칩n lineal m칰ltiple, los coeficientes $\beta_j$ se estiman utilizando el m칠todo de m칤nimos cuadrados ordinarios (OLS). Las estimaciones de los coeficientes se obtienen resolviendo el siguiente sistema de ecuaciones matriciales:

$\hat{\boldsymbol{\beta}} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$

donde:
- $\hat{\boldsymbol{\beta}}$ es el vector de coeficientes estimados $(\hat{\beta}_0, \hat{\beta}_1, \ldots, \hat{\beta}_p)^T$.
- $\mathbf{X}$ es la matriz de dise침o que incluye una columna de unos para el t칠rmino constante y las columnas de valores de las variables independientes.
- $\mathbf{y}$ es el vector de valores de la variable dependiente.

#### Varianza del Error

La varianza del error ($\sigma^2$) mide cu치n dispersos est치n los valores de $y$ alrededor de la l칤nea de regresi칩n. Se utiliza para cuantificar la variabilidad de los valores de $洧녽$ que no se explica por el modelo de regresi칩n lineal simple.

$\hat{\sigma}^2 = \frac{\sum (y_i - \hat{y}_i)^2}{n - p - 1}$

donde:
- $n$ es el n칰mero de observaciones.
- $p$ es el n칰mero de variables independientes.

#### Errores Est치ndar de los Coeficientes

Los errores est치ndar de los coeficientes estimados se calculan utilizando la matriz de varianza-covarianza de los coeficientes:

$\text{Var}(\hat{\boldsymbol{\beta}}) = \hat{\sigma}^2 (\mathbf{X}^T \mathbf{X})^{-1}$

El error est치ndar de cada coeficiente $\hat{\beta}_j$ es la ra칤z cuadrada del elemento correspondiente en la diagonal de la matriz de varianza-covarianza:

$\text{SE}(\hat{\beta}_j) = \sqrt{\text{Var}(\hat{\beta}_j)}$

El error est치ndar de un coeficiente mide la precisi칩n con la que se estima el coeficiente. Menor error estandar indica una mayor precision en la estimacion, mientras que un error alto indica una menor precisi칩n en la estimaci칩n


### Supuestos

Para que el modelo de regresi칩n lineal m칰ltiple sea v치lido, se deben cumplir ciertas asunciones:
1. **Linealidad:** La relaci칩n entre la variable dependiente y las variables independientes es lineal.
2. **Independencia:** Las observaciones son independientes entre s칤. En otras palabras, no autocorrelaci칩n.
3. **Homoscedasticidad:** La variabilidad de los errores es constante a lo largo de todos los valores de las variables independientes.
4. **Normalidad de los Errores:** Los errores de la regresi칩n son distribuidos normalmente.
5. **No Multicolinealidad:** Las variables independientes no est치n altamente correlacionadas entre s칤.

### Ejemplo de Aplicaci칩n

Supongamos que deseamos predecir el precio de venta de una casa ($y$) en funci칩n de su tama침o en pies cuadrados ($x_1$) y el n칰mero de habitaciones ($x_2$).

- **Datos de ejemplo:**
  $$
  \begin{array}{ccc}
  \text{Tama침o (pies cuadrados)} & \text{N칰mero de Habitaciones} & \text{Precio (miles de d칩lares)} \\
  1000 & 2 & 200 \\
  1500 & 3 & 250 \\
  2000 & 4 & 300 \\
  2500 & 3 & 350 \\
  3000 & 5 & 400 \\
  \end{array}
  $$

- **Ajuste del modelo:**
  Utilizando un software estad칤stico o una herramienta de programaci칩n como Python, se pueden ajustar los par치metros $\beta_0, \beta_1, \beta_2$ para obtener la ecuaci칩n de la regresi칩n m칰ltiple. Supongamos que obtenemos:
  $$ \text{Precio} = 50 + 0.05 \times \text{Tama침o} + 10 \times \text{N칰mero de Habitaciones} $$

En este caso, la intersecci칩n con el eje $y$ es 50 (que podr칤a interpretarse como el precio base en miles de d칩lares), la pendiente con respecto al tama침o es 0.05 (lo que indica que por cada incremento de 1 pie cuadrado en el tama침o de la casa, el precio aumenta en 0.05 miles de d칩lares), y la pendiente con respecto al n칰mero de habitaciones es 10 (lo que indica que por cada habitaci칩n adicional, el precio aumenta en 10 miles de d칩lares).

### Interpretaci칩n

La interpretaci칩n de los par치metros del modelo es crucial:
- **$\beta_0$ (intersecci칩n):** El valor esperado de $y$ cuando todas las variables independientes son 0. En el ejemplo, si el tama침o de la casa y el n칰mero de habitaciones fueran 0, el precio base ser칤a 50,000 d칩lares.
- **$\beta_1$ (pendiente para tama침o):** El cambio esperado en $y$ por cada unidad adicional de tama침o. En el ejemplo, cada pie cuadrado adicional se asocia con un incremento de 50 d칩lares en el precio.
- **$\beta_2$ (pendiente para n칰mero de habitaciones):** El cambio esperado en $y$ por cada habitaci칩n adicional. En el ejemplo, cada habitaci칩n adicional se asocia con un incremento de 10,000 d칩lares en el precio.

Este modelo permite analizar c칩mo varias caracter칤sticas de una propiedad influyen en su precio de venta de manera simult치nea.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
data = {
    'Tama침o': [1000, 1500, 2000, 2500, 3000],
    'N칰mero de Habitaciones': [2, 3, 4, 3, 5],
    'Precio': [200, 250, 300, 350, 400]
}

# Convertir a DataFrame
df = pd.DataFrame(data)

# Variables independientes y dependientes
X = df[['Tama침o', 'N칰mero de Habitaciones']]
y = df['Precio']

# Crear el modelo de regresi칩n lineal m칰ltiple
modelo = LinearRegression()

# Ajustar el modelo a los datos
modelo.fit(X, y)

# Obtener los par치metros del modelo
beta_0 = modelo.intercept_
beta_1, beta_2 = modelo.coef_

# Imprimir los par치metros
print(f"Intersecci칩n (0): {beta_0:.2f}")
print(f"Pendiente para Tama침o (1): {beta_1:.2f}")
print(f"Pendiente para N칰mero de Habitaciones (2): {beta_2:.2f}")

# Predicciones para visualizaci칩n
tama침o_pred = np.linspace(900, 3100, 10)
habitaciones_pred = np.linspace(1, 6, 10)
Tama침o_pred, Habitaciones_pred = np.meshgrid(tama침o_pred, habitaciones_pred)
X_pred = np.column_stack([Tama침o_pred.ravel(), Habitaciones_pred.ravel()])
precio_pred = modelo.predict(X_pred)
Precio_pred = precio_pred.reshape(Tama침o_pred.shape)

# Visualizaci칩n
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Tama침o'], df['N칰mero de Habitaciones'], df['Precio'], color='blue', label='Datos Observados')
ax.plot_surface(Tama침o_pred, Habitaciones_pred, Precio_pred, color='red', alpha=0.5, rstride=100, cstride=100)
ax.set_xlabel('Tama침o (pies cuadrados)')
ax.set_ylabel('N칰mero de Habitaciones')
ax.set_zlabel('Precio (miles de d칩lares)')
ax.set_title('Regresi칩n Lineal M칰ltiple')
ax.legend()
plt.show()
```

## Modelo de Regresi칩n Polinomial

El modelo de regresi칩n polinomial es una extensi칩n del modelo de regresi칩n lineal, utilizado para modelar la relaci칩n entre una variable dependiente (respuesta) y una o m치s variables independientes (predictoras) cuando esta relaci칩n es mejor capturada por una curva en lugar de una l칤nea recta. A diferencia de la regresi칩n lineal, en la regresi칩n polinomial se introducen t칠rminos polin칩micos de las variables independientes.

En la regresi칩n polinomial, la relaci칩n entre la variable dependiente $y$ y las variables independientes $x_1, x_2, \dots, x_n$ se modela como una combinaci칩n de t칠rminos polin칩micos de diferentes grados.

### Ecuaci칩n del Modelo

La ecuaci칩n general de un modelo de regresi칩n polinomial de grado $d$ para una 칰nica variable independiente $x$ es:

$
y = \beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3 + \dots + \beta_d x^d + \epsilon
$

donde:
- $y$ es la variable dependiente.
- $x$ es la variable independiente.
- $\beta_0, \beta_1, \dots, \beta_d$ son los coeficientes del modelo.
- $\epsilon$ es el t칠rmino de error, que captura la variabilidad en $y$ que no es explicada por el modelo.

Para m칰ltiples variables independientes $x_1, x_2, \dots, x_n$, la ecuaci칩n se extiende para incluir t칠rminos polin칩micos de cada variable, as칤 como interacciones entre ellas.

El prop칩sito del modelo de regresi칩n polinomial es capturar relaciones m치s complejas entre las variables que no pueden ser adecuadamente modeladas por una simple l칤nea recta. Al a침adir t칠rminos polin칩micos, el modelo se vuelve m치s flexible y puede ajustarse mejor a la forma real de los datos.

#### Estimaci칩n de Coeficientes

La estimaci칩n de los coeficientes $\beta_0, \beta_1, \dots, \beta_d$ se realiza utilizando el m칠todo de m칤nimos cuadrados ordinarios (OLS), al igual que en la regresi칩n lineal. Se buscan los valores de los coeficientes que minimicen la suma de los cuadrados de las diferencias entre los valores observados y los valores predichos por el modelo.

#### Ajuste del modelo

Para ajustar un modelo de regresi칩n polinomial a los datos, los pasos son similares a los de la regresi칩n lineal, pero con la diferencia de que se a침aden t칠rminos polin칩micos adicionales:

1. **Transformaci칩n de Variables**: Se transforman las variables independientes para incluir los t칠rminos polin칩micos.
2. **Estimaci칩n de Coeficientes**: Se estima los coeficientes utilizando el m칠todo de m칤nimos cuadrados.
3. **Evaluaci칩n del Modelo**: Se eval칰a el ajuste del modelo utilizando m칠tricas de desempe침o, como el coeficiente de determinaci칩n $R^2$ y el error cuadr치tico medio (MSE).

### Supuestos

Las asunciones del modelo de regresi칩n polinomial son similares a las de la regresi칩n lineal, con algunas consideraciones adicionales:
1. **Linealidad en los Coeficientes**: El modelo es lineal en los coeficientes $\beta$, aunque no en las variables $x$.
2. **Independencia de Observaciones**: Las observaciones deben ser independientes entre s칤.
3. **Homoscedasticidad**: La varianza de los errores debe ser constante.
4. **Normalidad de los Errores**: Los errores deben seguir una distribuci칩n normal.
5. **No Multicolinealidad**: Es importante verificar que los t칠rminos polin칩micos no est칠n altamente correlacionados entre s칤, lo cual podr칤a afectar la estabilidad del modelo.

### Ejemplo de Aplicaci칩n

Supongamos que deseamos predecir el precio de una casa ($y$) en funci칩n de su tama침o en pies cuadrados ($x$). Observamos que la relaci칩n entre el precio y el tama침o no es lineal, sino que sigue una curva, por lo que decidimos ajustar un modelo de regresi칩n polinomial de grado 2 (cuadr치tico).

- Datos de Ejemplo:

$
\begin{array}{cc}
\text{Tama침o (pies cuadrados)} & \text{Precio (miles de d칩lares)} \\
1000 & 150 \\
1500 & 200 \\
2000 & 260 \\
2500 & 350 \\
3000 & 470 \\
\end{array}
$

### Interpretaci칩n

- **Coeficientes $\beta_0, \beta_1, \beta_2$**: Los coeficientes estimados indican la contribuci칩n de cada t칠rmino (lineal y cuadr치tico) al modelo. El coeficiente de $x^2$ captura la curvatura de la relaci칩n entre el tama침o y el precio.
- **Intercepto $\beta_0$**: Representa el precio base de la casa cuando el tama침o es cero (lo cual, en este contexto, puede no tener una interpretaci칩n pr치ctica directa).


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Datos de ejemplo
X = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1, 1)
y = np.array([150, 200, 260, 350, 470])

# Transformaci칩n polinomial (grado 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Ajuste del modelo
model = LinearRegression()
model.fit(X_poly, y)

# Predicciones
y_pred = model.predict(X_poly)

# Coeficientes del modelo
print("Coeficientes:", model.coef_)
print("Intercepto:", model.intercept_)

# Visualizaci칩n del ajuste
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('Tama침o (pies cuadrados)')
plt.ylabel('Precio (miles de d칩lares)')
plt.title('Regresi칩n Polinomial')
plt.show()
```

## Supuestos de regresion lineal

Aqu칤 tienes diferentes m칠todos y t칠cnicas para probar y ajustar los datos para que cumplan con los supuestos de un modelo de regresi칩n lineal m칰ltiple:

### **Linealidad**

**M칠todos para probar:**
- **Gr치fica de Residuales vs. Predicciones:** Analiza el gr치fico de residuales para verificar si muestran una dispersi칩n aleatoria alrededor de cero. Una tendencia sistem치tica indica violaci칩n de linealidad.
- **Gr치ficas de Componentes y Residuos (CERES):** Estas gr치ficas ayudan a detectar relaciones no lineales entre las variables independientes y la dependiente.

**M칠todos para ajustar:**
- **Transformaciones de Variables:** Aplica transformaciones como logar칤tmica, cuadr치tica o polin칩mica a las variables independientes o dependientes para capturar relaciones no lineales.
- **Modelos No Lineales:** Considera el uso de modelos de regresi칩n no lineales si las transformaciones no son suficientes.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Datos de ejemplo
data = {
    'Tama침o': [1000, 1500, 2000, 2500, 3000],
    'N칰mero de Habitaciones': [2, 3, 4, 3, 5],
    'Precio': [200, 250, 300, 350, 400]
}
df = pd.DataFrame(data)
X = df[['Tama침o', 'N칰mero de Habitaciones']]
y = df['Precio']

# Crear y ajustar el modelo
modelo = LinearRegression()
modelo.fit(X, y)
y_pred = modelo.predict(X)

# Calcular residuos
residuos = y - y_pred

# Gr치fico de Residuos vs. Predicciones
plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuos, color='blue', edgecolor='k', alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicciones')
plt.ylabel('Residuos')
plt.title('Gr치fico de Residuos vs. Predicciones')
plt.grid(True)
plt.show()
```

### **Independencia**

**M칠todos para probar:**
- **Prueba de Durbin-Watson:** Esta prueba estad칤stica eval칰a la autocorrelaci칩n de los errores en un modelo de regresi칩n. Un valor cercano a 2 sugiere que no hay autocorrelaci칩n.
- **Gr치fica de Residuos:** Observa los residuos para detectar patrones que sugieran dependencia entre observaciones.

**M칠todos para ajustar:**
- **Modelos de Series Temporales:** Si los datos son de series temporales, utiliza modelos que manejen la autocorrelaci칩n como ARIMA.
- **A침adir Variables:** Incluye variables adicionales que puedan explicar la dependencia entre observaciones.



```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.stattools import durbin_watson

# Datos de ejemplo
data = {
    'Tama침o': [1000, 1500, 2000, 2500, 3000],
    'N칰mero de Habitaciones': [2, 3, 4, 3, 5],
    'Precio': [200, 250, 300, 350, 400]
}
df = pd.DataFrame(data)
X = df[['Tama침o', 'N칰mero de Habitaciones']]
y = df['Precio']

# Crear y ajustar el modelo
modelo = LinearRegression()
modelo.fit(X, y)
y_pred = modelo.predict(X)

# Calcular residuos
residuos = y - y_pred

# Prueba de Durbin-Watson
dw_stat = durbin_watson(residuos)
print(f"Estad칤stico de Durbin-Watson: {dw_stat:.2f}")
```

### **Homoscedasticidad**

**M칠todos para probar:**
- **Gr치fica de Residuos vs. Predicciones:** Analiza la dispersi칩n de los residuos. Un patr칩n en forma de embudo indica heteroscedasticidad.
- **Prueba de Breusch-Pagan:** Prueba estad칤stica que eval칰a si la varianza de los errores es constante.

**M칠todos para ajustar:**
- **Transformaciones de Variables:** Aplicar transformaciones a la variable dependiente (como logar칤tmica) puede ayudar a estabilizar la varianza.
- **Modelos de Regresi칩n Robustas:** Usa modelos robustos a la heteroscedasticidad como la regresi칩n robusta.



```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats import diagnostic as diag

# Datos de ejemplo
data = {
    'Tama침o': [1000, 1500, 2000, 2500, 3000],
    'N칰mero de Habitaciones': [2, 3, 4, 3, 5],
    'Precio': [200, 250, 300, 350, 400]
}
df = pd.DataFrame(data)
X = df[['Tama침o', 'N칰mero de Habitaciones']]
y = df['Precio']

# A침adir una constante para el modelo
X = sm.add_constant(X)

# Crear y ajustar el modelo
modelo = sm.OLS(y, X).fit()

# Prueba de Breusch-Pagan
_, pval, _, _ = diag.het_breuschpagan(modelo.resid, modelo.model.exog)
print(f"P-valor de la prueba de Breusch-Pagan: {pval:.2f}")
```

### **Normalidad de los Errores**

**M칠todos para probar:**
- **Histograma de Errores:** Visualiza el histograma de los errores para evaluar su distribuci칩n.
- **Gr치fico Q-Q (Quantile-Quantile):** Compara la distribuci칩n de los errores con una distribuci칩n normal te칩rica.
- **Prueba de Shapiro-Wilk o Kolmogorov-Smirnov:** Estas pruebas estad칤sticas eval칰an la normalidad de los errores.

**M칠todos para ajustar:**
- **Transformaciones de Variables:** Las transformaciones en la variable dependiente pueden ayudar a lograr normalidad en los errores.
- **M칠todos No Param칠tricos:** Considera el uso de m칠todos no param칠tricos si la normalidad no se puede lograr.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
data = {
    'Tama침o': [1000, 1500, 2000, 2500, 3000],
    'N칰mero de Habitaciones': [2, 3, 4, 3, 5],
    'Precio': [200, 250, 300, 350, 400]
}
df = pd.DataFrame(data)
X = df[['Tama침o', 'N칰mero de Habitaciones']]
y = df['Precio']

# Crear y ajustar el modelo
modelo = LinearRegression()
modelo.fit(X, y)
y_pred = modelo.predict(X)

# Calcular residuos
residuos = y - y_pred

# Gr치fico Q-Q
plt.figure(figsize=(10, 6))
stats.probplot(residuos, dist="norm", plot=plt)
plt.title('Gr치fico Q-Q de los Errores')
plt.grid(True)
plt.show()
```

### **No Multicolinealidad**

**M칠todos para probar:**
- **Matriz de Correlaci칩n:** Examina la matriz de correlaci칩n entre las variables independientes para detectar alta correlaci칩n.
- **칈ndice de Condici칩n y VIF (Factor de Inflaci칩n de la Varianza):** El VIF mide cu치nto se incrementa la varianza de los coeficientes debido a la colinealidad entre variables independientes. Un VIF alto indica que la variable est치 altamente correlacionada con otras variables independientes

$$VIF_j = \frac{1}{1-R_j^2}$$

donde $R_j^2$ es el $R^2$ del modelo de regresi칩n de la variable independiente $x_j$ contra las demas variables independientes

**M칠todos para ajustar:**
- **Eliminaci칩n de Variables:** Excluye variables altamente correlacionadas del modelo.
- **Combinaci칩n de Variables:** Combina variables correlacionadas en una sola variable (por ejemplo, mediante an치lisis de componentes principales).
- **Regularizaci칩n:** Usa t칠cnicas de regularizaci칩n como Lasso o Ridge que penalizan la multicolinealidad.

Estos m칠todos y t칠cnicas pueden ayudarte a validar y ajustar tus datos para cumplir con los supuestos de la regresi칩n lineal m칰ltiple, lo que contribuir치 a la precisi칩n y fiabilidad del modelo.

```python
# Ejemplo de multicolinearidad usnado VIF

import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Datos de ejemplo
data = {
    'Tama침o': [1000, 1500, 2000, 2500, 3000],
    'N칰mero de Habitaciones': [2, 3, 4, 3, 5],
    'Precio': [200, 250, 300, 350, 400]
}
df = pd.DataFrame(data)
X = df[['Tama침o', 'N칰mero de Habitaciones']]

# Calcular VIF
X_with_const = sm.add_constant(X)
vif_data = pd.DataFrame()
vif_data["Variable"] = X_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i) for i in range(X_with_const.shape[1])]

print(vif_data)
```

```python
# Ejemplo de multicolinearidad usando matrices de correlacion

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Datos de ejemplo
data = {
    'Tama침o': [1000, 1500, 2000, 2500, 3000],
    'N칰mero de Habitaciones': [2, 3, 4, 3, 5],
    'Edad de la Casa': [10, 15, 20, 15, 25],  # Nueva variable para demostrar multicolinealidad
    'Precio': [200, 250, 300, 350, 400]
}

# Convertir a DataFrame
df = pd.DataFrame(data)

# Calcular la matriz de correlaci칩n
corr_matrix = df.corr()

# Mostrar la matriz de correlaci칩n
print("Matriz de Correlaci칩n:")
print(corr_matrix)

# Visualizar la matriz de correlaci칩n
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlaci칩n')
plt.show()
```
