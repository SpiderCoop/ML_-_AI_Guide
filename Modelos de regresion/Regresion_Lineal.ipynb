# Modelos de Regresión Lineal

Los modelos de regresión son herramientas fundamentales en estadística y machine learning, utilizadas para predecir una variable dependiente (respuesta) en función de una o más variables independientes (predictoras). Aquí se detallan algunas de las utilidades clave de los modelos de regresión:

- **Predicción:** Predicen valores continuos como precios de casas, ingresos anuales y temperaturas. Útiles para pronósticos futuros en finanzas, economía y climatología.

- **Análisis de Relaciones:** Ayudan a entender la relación entre variables dependientes e independientes, e identificar los factores más importantes que influyen en la variable de interés.

- **Optimización y Toma de Decisiones:** Utilizados para optimizar procesos industriales y de negocios, y proveer una base cuantitativa para decisiones informadas como presupuestos de marketing.

- **Detección de Tendencias y Patrones:** Detectan tendencias a lo largo del tiempo y ayudan a identificar anomalías que pueden indicar problemas u oportunidades.

- **Evaluación de Impacto:** Evalúan el impacto de políticas públicas o decisiones de negocios, y miden el efecto de intervenciones específicas como campañas de marketing.

- **Análisis Comparativo:** Comparan el efecto de diferentes grupos o tratamientos, y permiten el benchmarking para comparar el rendimiento con estándares de la industria.

## Modelo de Regresión Lineal Simple

El modelo de regresión lineal simple es una técnica estadística utilizada para modelar la relación entre dos variables: una variable dependiente (respuesta) y una variable independiente (predictora). La relación entre estas variables se representa mediante una línea recta.

### Ecuación del Modelo

La ecuación del modelo de regresión lineal simple es:

$ y = \beta_0 + \beta_1 x + \epsilon $

donde:
- $ y $ es la variable dependiente.
- $ x $ es la variable independiente.
- $ \beta_0 $ es la intersección con el eje $ y $ (constante o término independiente).
- $ \beta_1 $ es la pendiente de la línea de regresión, que representa el cambio en $ y $ por unidad de cambio en $ x $.
- $ \epsilon $ es el término de error, que captura la variabilidad en $ y $ que no se explica por la relación lineal con $ x $.

El propósito del modelo de regresión lineal simple es encontrar los valores de $ \beta_0 $ y $ \beta_1 $ que minimicen la suma de los cuadrados de las diferencias entre los valores observados y los valores predichos por el modelo. Este proceso se conoce como ajuste de mínimos cuadrados ordinarios.

#### Estimacion de coeficientes

$\hat{\beta}_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}$

$\hat{\beta}_0 = \bar{y} - \hat{\beta}_1 \bar{x}$

#### Varianza del error

La varianza del error ($\sigma^2$) mide cuán dispersos están los valores de $y$ alrededor de la línea de regresión. Se utiliza para cuantificar la variabilidad de los valores de $𝑦$ que no se explica por el modelo de regresión lineal simple.

$\sigma^2 = \frac{\sum (y_i - \hat{y}_i)^2}{n - 2}$

#### Error estandar

El error estándar de un coeficiente mide la precisión con la que se estima el coeficiente. Menor error estandar indica una mayor precision en la estimacion, mientras que un error alto indica una menor precisión en la estimación

$\text{SE}(\hat{\beta}_1) = \sqrt{\frac{\sigma^2}{\sum (x_i - \bar{x})^2}}$

$\text{SE}(\hat{\beta}_0) = \sqrt{\sigma^2 \left( \frac{1}{n} + \frac{\bar{x}^2}{\sum (x_i - \bar{x})^2} \right)}$


### Supuestos

Para que el modelo de regresión lineal simple sea válido, se deben cumplir ciertas asunciones:
1. **Linealidad:** La relación entre la variable dependiente y la variable independiente es lineal.
2. **Independencia:** Las observaciones son independientes entre sí. En otras palabras, que no exista autocorrelación
3. **Homoscedasticidad:** La variabilidad de los errores es constante a lo largo de todos los valores de la variable independiente.
4. **Normalidad de los Errores:** Los errores de la regresión son distribuidos normalmente.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
tamaño = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1, 1)  # Convertimos a formato de columna
precio = np.array([200, 250, 300, 350, 400])

# Crear el modelo de regresión lineal
modelo = LinearRegression()

# Ajustar el modelo a los datos
modelo.fit(tamaño, precio)

# Obtener los parámetros del modelo
beta_0 = modelo.intercept_
beta_1 = modelo.coef_[0]

# Imprimir los parámetros
print(f"Intersección (β0): {beta_0:.2f}")
print(f"Pendiente (β1): {beta_1:.2f}")

# Predicciones
tamaño_pred = np.linspace(900, 3100, 100).reshape(-1, 1)
precio_pred = modelo.predict(tamaño_pred)

# Visualización
plt.figure(figsize=(10, 6))
plt.scatter(tamaño, precio, color='blue', label='Datos Observados')
plt.plot(tamaño_pred, precio_pred, color='red', linewidth=2, label='Línea de Regresión')
plt.xlabel('Tamaño (pies cuadrados)')
plt.ylabel('Precio (miles de dólares)')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.grid(True)
plt.show()
```

## Modelo de Regresión Lineal Múltiple

El modelo de regresión lineal múltiple es una extensión del modelo de regresión lineal simple que se utiliza para modelar la relación entre una variable dependiente y dos o más variables independientes. Este modelo permite examinar cómo varias variables predictoras influyen en una variable respuesta.

### Ecuación del Modelo

La ecuación del modelo de regresión lineal múltiple es:

$$ y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_n x_n + \epsilon $$

donde:
- $y$ es la variable dependiente.
- $x_1, x_2, \ldots, x_n$ son las variables independientes (predictoras).
- $\beta_0$ es la intersección con el eje $y$ (constante o término independiente).
- $\beta_1, \beta_2, \ldots, \beta_n$ son los coeficientes de las variables independientes, que representan el cambio en $y$ por unidad de cambio en cada $x_i$.
- $\epsilon$ es el término de error, que captura la variabilidad en $y$ que no se explica por la relación lineal con las $x_i$.

El propósito del modelo de regresión lineal múltiple es encontrar los valores de $\beta_0, \beta_1, \ldots, \beta_n$ que minimicen la suma de los cuadrados de las diferencias entre los valores observados y los valores predichos por el modelo. Este proceso se realiza utilizando el ajuste de mínimos cuadrados ordinarios.

#### Estimacion de coeficientes 

En regresión lineal múltiple, los coeficientes $\beta_j$ se estiman utilizando el método de mínimos cuadrados ordinarios (OLS). Las estimaciones de los coeficientes se obtienen resolviendo el siguiente sistema de ecuaciones matriciales:

$\hat{\boldsymbol{\beta}} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$

donde:
- $\hat{\boldsymbol{\beta}}$ es el vector de coeficientes estimados $(\hat{\beta}_0, \hat{\beta}_1, \ldots, \hat{\beta}_p)^T$.
- $\mathbf{X}$ es la matriz de diseño que incluye una columna de unos para el término constante y las columnas de valores de las variables independientes.
- $\mathbf{y}$ es el vector de valores de la variable dependiente.

#### Varianza del Error

La varianza del error ($\sigma^2$) mide cuán dispersos están los valores de $y$ alrededor de la línea de regresión. Se utiliza para cuantificar la variabilidad de los valores de $𝑦$ que no se explica por el modelo de regresión lineal simple.

$\hat{\sigma}^2 = \frac{\sum (y_i - \hat{y}_i)^2}{n - p - 1}$

donde:
- $n$ es el número de observaciones.
- $p$ es el número de variables independientes.

#### Errores Estándar de los Coeficientes

Los errores estándar de los coeficientes estimados se calculan utilizando la matriz de varianza-covarianza de los coeficientes:

$\text{Var}(\hat{\boldsymbol{\beta}}) = \hat{\sigma}^2 (\mathbf{X}^T \mathbf{X})^{-1}$

El error estándar de cada coeficiente $\hat{\beta}_j$ es la raíz cuadrada del elemento correspondiente en la diagonal de la matriz de varianza-covarianza:

$\text{SE}(\hat{\beta}_j) = \sqrt{\text{Var}(\hat{\beta}_j)}$

El error estándar de un coeficiente mide la precisión con la que se estima el coeficiente. Menor error estandar indica una mayor precision en la estimacion, mientras que un error alto indica una menor precisión en la estimación


### Supuestos

Para que el modelo de regresión lineal múltiple sea válido, se deben cumplir ciertas asunciones:
1. **Linealidad:** La relación entre la variable dependiente y las variables independientes es lineal.
2. **Independencia:** Las observaciones son independientes entre sí. En otras palabras, no autocorrelación.
3. **Homoscedasticidad:** La variabilidad de los errores es constante a lo largo de todos los valores de las variables independientes.
4. **Normalidad de los Errores:** Los errores de la regresión son distribuidos normalmente.
5. **No Multicolinealidad:** Las variables independientes no están altamente correlacionadas entre sí.

### Ejemplo de Aplicación

Supongamos que deseamos predecir el precio de venta de una casa ($y$) en función de su tamaño en pies cuadrados ($x_1$) y el número de habitaciones ($x_2$).

- **Datos de ejemplo:**
  $$
  \begin{array}{ccc}
  \text{Tamaño (pies cuadrados)} & \text{Número de Habitaciones} & \text{Precio (miles de dólares)} \\
  1000 & 2 & 200 \\
  1500 & 3 & 250 \\
  2000 & 4 & 300 \\
  2500 & 3 & 350 \\
  3000 & 5 & 400 \\
  \end{array}
  $$

- **Ajuste del modelo:**
  Utilizando un software estadístico o una herramienta de programación como Python, se pueden ajustar los parámetros $\beta_0, \beta_1, \beta_2$ para obtener la ecuación de la regresión múltiple. Supongamos que obtenemos:
  $$ \text{Precio} = 50 + 0.05 \times \text{Tamaño} + 10 \times \text{Número de Habitaciones} $$

En este caso, la intersección con el eje $y$ es 50 (que podría interpretarse como el precio base en miles de dólares), la pendiente con respecto al tamaño es 0.05 (lo que indica que por cada incremento de 1 pie cuadrado en el tamaño de la casa, el precio aumenta en 0.05 miles de dólares), y la pendiente con respecto al número de habitaciones es 10 (lo que indica que por cada habitación adicional, el precio aumenta en 10 miles de dólares).

### Interpretación

La interpretación de los parámetros del modelo es crucial:
- **$\beta_0$ (intersección):** El valor esperado de $y$ cuando todas las variables independientes son 0. En el ejemplo, si el tamaño de la casa y el número de habitaciones fueran 0, el precio base sería 50,000 dólares.
- **$\beta_1$ (pendiente para tamaño):** El cambio esperado en $y$ por cada unidad adicional de tamaño. En el ejemplo, cada pie cuadrado adicional se asocia con un incremento de 50 dólares en el precio.
- **$\beta_2$ (pendiente para número de habitaciones):** El cambio esperado en $y$ por cada habitación adicional. En el ejemplo, cada habitación adicional se asocia con un incremento de 10,000 dólares en el precio.

Este modelo permite analizar cómo varias características de una propiedad influyen en su precio de venta de manera simultánea.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
data = {
    'Tamaño': [1000, 1500, 2000, 2500, 3000],
    'Número de Habitaciones': [2, 3, 4, 3, 5],
    'Precio': [200, 250, 300, 350, 400]
}

# Convertir a DataFrame
df = pd.DataFrame(data)

# Variables independientes y dependientes
X = df[['Tamaño', 'Número de Habitaciones']]
y = df['Precio']

# Crear el modelo de regresión lineal múltiple
modelo = LinearRegression()

# Ajustar el modelo a los datos
modelo.fit(X, y)

# Obtener los parámetros del modelo
beta_0 = modelo.intercept_
beta_1, beta_2 = modelo.coef_

# Imprimir los parámetros
print(f"Intersección (β0): {beta_0:.2f}")
print(f"Pendiente para Tamaño (β1): {beta_1:.2f}")
print(f"Pendiente para Número de Habitaciones (β2): {beta_2:.2f}")

# Predicciones para visualización
tamaño_pred = np.linspace(900, 3100, 10)
habitaciones_pred = np.linspace(1, 6, 10)
Tamaño_pred, Habitaciones_pred = np.meshgrid(tamaño_pred, habitaciones_pred)
X_pred = np.column_stack([Tamaño_pred.ravel(), Habitaciones_pred.ravel()])
precio_pred = modelo.predict(X_pred)
Precio_pred = precio_pred.reshape(Tamaño_pred.shape)

# Visualización
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Tamaño'], df['Número de Habitaciones'], df['Precio'], color='blue', label='Datos Observados')
ax.plot_surface(Tamaño_pred, Habitaciones_pred, Precio_pred, color='red', alpha=0.5, rstride=100, cstride=100)
ax.set_xlabel('Tamaño (pies cuadrados)')
ax.set_ylabel('Número de Habitaciones')
ax.set_zlabel('Precio (miles de dólares)')
ax.set_title('Regresión Lineal Múltiple')
ax.legend()
plt.show()
```

## Modelo de Regresión Polinomial

El modelo de regresión polinomial es una extensión del modelo de regresión lineal, utilizado para modelar la relación entre una variable dependiente (respuesta) y una o más variables independientes (predictoras) cuando esta relación es mejor capturada por una curva en lugar de una línea recta. A diferencia de la regresión lineal, en la regresión polinomial se introducen términos polinómicos de las variables independientes.

En la regresión polinomial, la relación entre la variable dependiente $y$ y las variables independientes $x_1, x_2, \dots, x_n$ se modela como una combinación de términos polinómicos de diferentes grados.

### Ecuación del Modelo

La ecuación general de un modelo de regresión polinomial de grado $d$ para una única variable independiente $x$ es:

$
y = \beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3 + \dots + \beta_d x^d + \epsilon
$

donde:
- $y$ es la variable dependiente.
- $x$ es la variable independiente.
- $\beta_0, \beta_1, \dots, \beta_d$ son los coeficientes del modelo.
- $\epsilon$ es el término de error, que captura la variabilidad en $y$ que no es explicada por el modelo.

Para múltiples variables independientes $x_1, x_2, \dots, x_n$, la ecuación se extiende para incluir términos polinómicos de cada variable, así como interacciones entre ellas.

El propósito del modelo de regresión polinomial es capturar relaciones más complejas entre las variables que no pueden ser adecuadamente modeladas por una simple línea recta. Al añadir términos polinómicos, el modelo se vuelve más flexible y puede ajustarse mejor a la forma real de los datos.

#### Estimación de Coeficientes

La estimación de los coeficientes $\beta_0, \beta_1, \dots, \beta_d$ se realiza utilizando el método de mínimos cuadrados ordinarios (OLS), al igual que en la regresión lineal. Se buscan los valores de los coeficientes que minimicen la suma de los cuadrados de las diferencias entre los valores observados y los valores predichos por el modelo.

#### Ajuste del modelo

Para ajustar un modelo de regresión polinomial a los datos, los pasos son similares a los de la regresión lineal, pero con la diferencia de que se añaden términos polinómicos adicionales:

1. **Transformación de Variables**: Se transforman las variables independientes para incluir los términos polinómicos.
2. **Estimación de Coeficientes**: Se estima los coeficientes utilizando el método de mínimos cuadrados.
3. **Evaluación del Modelo**: Se evalúa el ajuste del modelo utilizando métricas de desempeño, como el coeficiente de determinación $R^2$ y el error cuadrático medio (MSE).

### Supuestos

Las asunciones del modelo de regresión polinomial son similares a las de la regresión lineal, con algunas consideraciones adicionales:
1. **Linealidad en los Coeficientes**: El modelo es lineal en los coeficientes $\beta$, aunque no en las variables $x$.
2. **Independencia de Observaciones**: Las observaciones deben ser independientes entre sí.
3. **Homoscedasticidad**: La varianza de los errores debe ser constante.
4. **Normalidad de los Errores**: Los errores deben seguir una distribución normal.
5. **No Multicolinealidad**: Es importante verificar que los términos polinómicos no estén altamente correlacionados entre sí, lo cual podría afectar la estabilidad del modelo.

### Ejemplo de Aplicación

Supongamos que deseamos predecir el precio de una casa ($y$) en función de su tamaño en pies cuadrados ($x$). Observamos que la relación entre el precio y el tamaño no es lineal, sino que sigue una curva, por lo que decidimos ajustar un modelo de regresión polinomial de grado 2 (cuadrático).

- Datos de Ejemplo:

$
\begin{array}{cc}
\text{Tamaño (pies cuadrados)} & \text{Precio (miles de dólares)} \\
1000 & 150 \\
1500 & 200 \\
2000 & 260 \\
2500 & 350 \\
3000 & 470 \\
\end{array}
$

### Interpretación

- **Coeficientes $\beta_0, \beta_1, \beta_2$**: Los coeficientes estimados indican la contribución de cada término (lineal y cuadrático) al modelo. El coeficiente de $x^2$ captura la curvatura de la relación entre el tamaño y el precio.
- **Intercepto $\beta_0$**: Representa el precio base de la casa cuando el tamaño es cero (lo cual, en este contexto, puede no tener una interpretación práctica directa).


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Datos de ejemplo
X = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1, 1)
y = np.array([150, 200, 260, 350, 470])

# Transformación polinomial (grado 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Ajuste del modelo
model = LinearRegression()
model.fit(X_poly, y)

# Predicciones
y_pred = model.predict(X_poly)

# Coeficientes del modelo
print("Coeficientes:", model.coef_)
print("Intercepto:", model.intercept_)

# Visualización del ajuste
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('Tamaño (pies cuadrados)')
plt.ylabel('Precio (miles de dólares)')
plt.title('Regresión Polinomial')
plt.show()
```

## Supuestos de regresion lineal

Aquí tienes diferentes métodos y técnicas para probar y ajustar los datos para que cumplan con los supuestos de un modelo de regresión lineal múltiple:

### **Linealidad**

**Métodos para probar:**
- **Gráfica de Residuales vs. Predicciones:** Analiza el gráfico de residuales para verificar si muestran una dispersión aleatoria alrededor de cero. Una tendencia sistemática indica violación de linealidad.
- **Gráficas de Componentes y Residuos (CERES):** Estas gráficas ayudan a detectar relaciones no lineales entre las variables independientes y la dependiente.

**Métodos para ajustar:**
- **Transformaciones de Variables:** Aplica transformaciones como logarítmica, cuadrática o polinómica a las variables independientes o dependientes para capturar relaciones no lineales.
- **Modelos No Lineales:** Considera el uso de modelos de regresión no lineales si las transformaciones no son suficientes.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Datos de ejemplo
data = {
    'Tamaño': [1000, 1500, 2000, 2500, 3000],
    'Número de Habitaciones': [2, 3, 4, 3, 5],
    'Precio': [200, 250, 300, 350, 400]
}
df = pd.DataFrame(data)
X = df[['Tamaño', 'Número de Habitaciones']]
y = df['Precio']

# Crear y ajustar el modelo
modelo = LinearRegression()
modelo.fit(X, y)
y_pred = modelo.predict(X)

# Calcular residuos
residuos = y - y_pred

# Gráfico de Residuos vs. Predicciones
plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuos, color='blue', edgecolor='k', alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicciones')
plt.ylabel('Residuos')
plt.title('Gráfico de Residuos vs. Predicciones')
plt.grid(True)
plt.show()
```

### **Independencia**

**Métodos para probar:**
- **Prueba de Durbin-Watson:** Esta prueba estadística evalúa la autocorrelación de los errores en un modelo de regresión. Un valor cercano a 2 sugiere que no hay autocorrelación.
- **Gráfica de Residuos:** Observa los residuos para detectar patrones que sugieran dependencia entre observaciones.

**Métodos para ajustar:**
- **Modelos de Series Temporales:** Si los datos son de series temporales, utiliza modelos que manejen la autocorrelación como ARIMA.
- **Añadir Variables:** Incluye variables adicionales que puedan explicar la dependencia entre observaciones.



```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.stattools import durbin_watson

# Datos de ejemplo
data = {
    'Tamaño': [1000, 1500, 2000, 2500, 3000],
    'Número de Habitaciones': [2, 3, 4, 3, 5],
    'Precio': [200, 250, 300, 350, 400]
}
df = pd.DataFrame(data)
X = df[['Tamaño', 'Número de Habitaciones']]
y = df['Precio']

# Crear y ajustar el modelo
modelo = LinearRegression()
modelo.fit(X, y)
y_pred = modelo.predict(X)

# Calcular residuos
residuos = y - y_pred

# Prueba de Durbin-Watson
dw_stat = durbin_watson(residuos)
print(f"Estadístico de Durbin-Watson: {dw_stat:.2f}")
```

### **Homoscedasticidad**

**Métodos para probar:**
- **Gráfica de Residuos vs. Predicciones:** Analiza la dispersión de los residuos. Un patrón en forma de embudo indica heteroscedasticidad.
- **Prueba de Breusch-Pagan:** Prueba estadística que evalúa si la varianza de los errores es constante.

**Métodos para ajustar:**
- **Transformaciones de Variables:** Aplicar transformaciones a la variable dependiente (como logarítmica) puede ayudar a estabilizar la varianza.
- **Modelos de Regresión Robustas:** Usa modelos robustos a la heteroscedasticidad como la regresión robusta.



```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats import diagnostic as diag

# Datos de ejemplo
data = {
    'Tamaño': [1000, 1500, 2000, 2500, 3000],
    'Número de Habitaciones': [2, 3, 4, 3, 5],
    'Precio': [200, 250, 300, 350, 400]
}
df = pd.DataFrame(data)
X = df[['Tamaño', 'Número de Habitaciones']]
y = df['Precio']

# Añadir una constante para el modelo
X = sm.add_constant(X)

# Crear y ajustar el modelo
modelo = sm.OLS(y, X).fit()

# Prueba de Breusch-Pagan
_, pval, _, _ = diag.het_breuschpagan(modelo.resid, modelo.model.exog)
print(f"P-valor de la prueba de Breusch-Pagan: {pval:.2f}")
```

### **Normalidad de los Errores**

**Métodos para probar:**
- **Histograma de Errores:** Visualiza el histograma de los errores para evaluar su distribución.
- **Gráfico Q-Q (Quantile-Quantile):** Compara la distribución de los errores con una distribución normal teórica.
- **Prueba de Shapiro-Wilk o Kolmogorov-Smirnov:** Estas pruebas estadísticas evalúan la normalidad de los errores.

**Métodos para ajustar:**
- **Transformaciones de Variables:** Las transformaciones en la variable dependiente pueden ayudar a lograr normalidad en los errores.
- **Métodos No Paramétricos:** Considera el uso de métodos no paramétricos si la normalidad no se puede lograr.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
data = {
    'Tamaño': [1000, 1500, 2000, 2500, 3000],
    'Número de Habitaciones': [2, 3, 4, 3, 5],
    'Precio': [200, 250, 300, 350, 400]
}
df = pd.DataFrame(data)
X = df[['Tamaño', 'Número de Habitaciones']]
y = df['Precio']

# Crear y ajustar el modelo
modelo = LinearRegression()
modelo.fit(X, y)
y_pred = modelo.predict(X)

# Calcular residuos
residuos = y - y_pred

# Gráfico Q-Q
plt.figure(figsize=(10, 6))
stats.probplot(residuos, dist="norm", plot=plt)
plt.title('Gráfico Q-Q de los Errores')
plt.grid(True)
plt.show()
```

### **No Multicolinealidad**

**Métodos para probar:**
- **Matriz de Correlación:** Examina la matriz de correlación entre las variables independientes para detectar alta correlación.
- **Índice de Condición y VIF (Factor de Inflación de la Varianza):** El VIF mide cuánto se incrementa la varianza de los coeficientes debido a la colinealidad entre variables independientes. Un VIF alto indica que la variable está altamente correlacionada con otras variables independientes

$$VIF_j = \frac{1}{1-R_j^2}$$

donde $R_j^2$ es el $R^2$ del modelo de regresión de la variable independiente $x_j$ contra las demas variables independientes

**Métodos para ajustar:**
- **Eliminación de Variables:** Excluye variables altamente correlacionadas del modelo.
- **Combinación de Variables:** Combina variables correlacionadas en una sola variable (por ejemplo, mediante análisis de componentes principales).
- **Regularización:** Usa técnicas de regularización como Lasso o Ridge que penalizan la multicolinealidad.

Estos métodos y técnicas pueden ayudarte a validar y ajustar tus datos para cumplir con los supuestos de la regresión lineal múltiple, lo que contribuirá a la precisión y fiabilidad del modelo.

```python
# Ejemplo de multicolinearidad usando matrices de correlacion

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Datos de ejemplo
data = {
    'Tamaño': [1000, 1500, 2000, 2500, 3000],
    'Número de Habitaciones': [2, 3, 4, 3, 5],
    'Edad de la Casa': [10, 15, 20, 15, 25],  # Nueva variable para demostrar multicolinealidad
    'Precio': [200, 250, 300, 350, 400]
}

# Convertir a DataFrame
df = pd.DataFrame(data)

# Calcular la matriz de correlación
corr_matrix = df.corr()

# Mostrar la matriz de correlación
print("Matriz de Correlación:")
print(corr_matrix)

# Visualizar la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlación')
plt.show()
```
